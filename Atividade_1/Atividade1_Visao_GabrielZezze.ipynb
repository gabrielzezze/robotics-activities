# Equalização de histograma

```python
rintin = cv2.imread("images/RinTinTin.jpg")
```

```python
rintin.shape
```

```python
rintin_gray = cv2.cvtColor(rintin, cv2.COLOR_RGB2GRAY)
```

```python
rintin_gray.shape
```

```python
plt.imshow(rintin_gray, cmap="Greys_r", vmin=0, vmax=255)
```

```python

hist,bins = np.histogram(rintin_gray.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(rintin_gray.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
print("Max: {}".format(rintin_gray.max()))
print("Min: {}".format(rintin_gray.min()))
```

## Atividade 1

Realize as seguintes atividades:

1. O histograma fornecido aproveita bem as faixas de cores? 
2. Equalize o histograma de maneira que a cumulativa seja uma função linear na diagonal do gráfico
3. Plote a imagem desenhada

# 1)
#### O histograma fornecido não aproveita bem as faixas de cores pois no histograma acima pode ser observado que os pixel se encontra entre uma faixa de 8 a 58 enquanto é possivel utilizar uma faixa de 0 a 255.

```python
img2 = rintin_gray.copy()
# Funcao Reta #
def F(x):
    return (255*x)/50 - 40.8
# Substituicao de pixeis na copia da imagem #
for i in range(len(rintin_gray)):
    for x in range(len(rintin_gray[i])):
        pixel = F(rintin_gray[i][x])
        img2[i][x] = pixel
```

```python
# CDF #
hist,bins = np.histogram(img2.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()
# Hist Plot #
plt.plot(cdf_normalized, color = 'b')
plt.hist(img2.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.title("Histograma com pixeis ajustados")
plt.show()
```

```python
plt.imshow(img2,cmap="Greys_r", vmin=0, vmax=255)
plt.title("Imagem com pixeis ajustados")
plt.show()
```

Crie uma cópia da versão Grayscale figura acima, em que cada pixel  recebe a subtração entre o pixel que vem depois e o que vem antes dele. Apresente a imagem resultado.

# 3)

```python
image3 = cv2.imread('images/hall_box_battery_atividade3.png')
imgGray = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)
img2 = imgGray.copy()

height,width = imgGray.shape
for i in range(0,height):
    for x in range(1,width-1):
        img2[i][x] = abs(int(imgGray[i][x+1]) - int(imgGray[i][x-1]))       
```
